# 项目报告书
## 基于量化金融因子的实时因子分析与智能股市预测系统

## 一、系统架构

app:这个项目的后端。

front-endv2:这个项目的前端。我们使用vue3来构建前端，使用电子图表作为图表库。

我们使用PostgreSQL作为本项目的数据库。

## 二、功能展示

### 本项目分为五个模块：算法模块、策略模块、数据模块、因子分析和上传数据

### 1.算法模块

#### Step1：选择数据集

#### Step2：选择算法模型
我们目前提供了传统线性回归、xgboost、VAR、PatchTST模型来做预测，后续还会添加更多模型。

选择可选日期进行预测，没有数据的日期不可选择。

选择了模型与日期后，将会绘制出当前日期前25天的历史数据，以及后五天的实际数据与预测数据，并给出评价算法的指标，MSE与MAE的值。

### 2.策略模块

#### Step1：依据模型预测的结果，选择选股策略后，会显示模型参数，并绘制出预测时间范围内收益率的折线图。

#### Step2：选择可选日期后，可以显示出当日排名前五的五只股票代码。

### 3、数据模块

#### Step1：选择需要可视化的数据集

#### Step2：选择图表类型。我们目前支持三种图表类型，后续还会添加更多的图表组件。

##### 1、异常检测图（TSL）

数据集格式要求如下：

    | date | y   | y_pred |
    | ---- | --- | -----  |
    | 1    | 1.  | 1.2    |
    | 2    | 1.1 | 1.3    |
    
' date '为时间戳，' y '为时间戳的实际值，' y_pred '为时间戳的预测值。

##### 2、时序预测图（LPC）

数据集格式要求如下：

    | date | xx_real | xx_predict | [can be other property] |
    | ---- | ------- | ---------- | ----------------------- |
    | 1    | 1       | null       |                         |
    | 2    | null    | 2          |                         |
    
' date '为时间戳，' xx_real '为时间戳的实际值，' xx_predict '为时间戳的预测值。
可以在表中添加其他属性，绘图时将忽略。

##### 3、表格（Table）

显示数据集中的某几列数据，以表格的形式展现。

#### Step3：选择列名，绘制出所需的图像

### 4、因子分析

#### 本模块选择策略后，会做出对应的sharpe值图像。如果后续算法创新出有关因子权重的选择策略，则还会画出因子权重的走势图。

### 5、上传数据

#### 本模块可供用户上传自定义数据到我们的数据库中，再对其进行其它模块的分析操作。
